````python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
def plot_numerical_distribution(df, columns):
    '''
    –í—ã–≤–æ–¥–∏—Ç —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö —É–∫–∞–∑–∞–Ω–Ω–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ –≤ –≤–∏–¥–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –∏ –±–æ–∫—Å–ø–ª–æ—Ç
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - df: –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
    - columns: –≤—ã–±—Ä–∞–Ω–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü –¥–ª—è –≤—ã–≤–æ–¥–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
    '''
    for col in columns:
        # –°–æ–∑–¥–∞–µ–º —Ñ–∏–≥—É—Ä—É —Å –¥–≤—É–º—è –≥—Ä–∞—Ñ–∏–∫–∞–º–∏
        fig, axes = plt.subplots(2, 1, figsize=(8, 5), gridspec_kw={'height_ratios': [3, 1]}, constrained_layout=True)
        
        # –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
        sns.histplot(df[col].dropna(), bins=30, kde=False, ax=axes[0], color='blue', edgecolor='black')
        axes[0].set_title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ {col}', fontsize=12, fontweight='bold')
        axes[0].set_xlabel(None)
        axes[0].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=10)
        axes[0].grid(axis='y', linestyle='--', alpha=0.5)
        
        # –î–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞–∑–º–∞—Ö–∞ (–±–æ–∫—Å–ø–ª–æ—Ç)
        sns.boxplot(x=df[col].dropna(), ax=axes[1], color='blue')
        axes[1].set_xlabel(col, fontsize=10)
        
        # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤
        plt.show()
        
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏, –µ—Å—Ç—å –ª–∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –¥–æ –ø–µ—Ä–≤–æ–≥–æ –∑–∞–º–µ—Ä–∞ –∏–ª–∏ –ø–æ—Å–ª–µ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ
def check_timing(row, time_columns, first_measurement_col, last_measurement_col):
    '''
    –ü—Ä–æ–≤–µ—Ä—è–µ—Ç, –µ—Å—Ç—å –ª–∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –º–∞—Ç–µ—Ä–∏–∞–ª–æ–≤ –¥–æ –ø–µ—Ä–≤–æ–≥–æ –∑–∞–º–µ—Ä–∞ –∏–ª–∏ –ø–æ—Å–ª–µ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ –∑–∞–º–µ—Ä–∞.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - row: —Å—Ç—Ä–æ–∫–∞ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞.
    - time_columns: —Å–ø–∏—Å–æ–∫ —Å—Ç–æ–ª–±—Ü–æ–≤ —Å –≤—Ä–µ–º–µ–Ω–µ–º –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –º–∞—Ç–µ—Ä–∏–∞–ª–æ–≤.
    - first_measurement_col: –Ω–∞–∑–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ —Å –≤—Ä–µ–º–µ–Ω–µ–º –ø–µ—Ä–≤–æ–≥–æ –∑–∞–º–µ—Ä–∞.
    - last_measurement_col: –Ω–∞–∑–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ —Å –≤—Ä–µ–º–µ–Ω–µ–º –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ –∑–∞–º–µ—Ä–∞.

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    - True, –µ—Å–ª–∏ –µ—Å—Ç—å –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º—ã–µ –¥–æ–±–∞–≤–ª–µ–Ω–∏—è, –∏–Ω–∞—á–µ False.
    '''
    for col in time_columns:
        material_time = row[col]
        if pd.notna(material_time):
            if material_time < row[first_measurement_col]:
                return True
            if material_time > row[last_measurement_col]:
                return True
    return False
````

#### data_temp ‚Äî —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã

–ö–∞–∫ –≤–∏–¥–∏–º –∏–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, –µ—Å—Ç—å –∞–Ω–æ–º–∞–ª—å–Ω—ã–µ –Ω–∏–∑–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –æ–∫–æ–ª–æ 1200 –≥—Ä–∞–¥—É—Å–æ–≤. –£–¥–∞–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∏–∂–µ —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã –ø–ª–∞–≤–ª–µ–Ω–∏—è - 1500 –≥—Ä–∞–¥—É—Å–æ–≤.

````python
# –£–¥–∞–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã –Ω–∏–∂–µ 1500
data_temp = data_temp[data_temp['temp'] >= 1500]

# –£–¥–∞–ª–µ–Ω–∏–µ –∫–ª—é—á–µ–π —Å —Ç–æ–ª—å–∫–æ –æ–¥–Ω–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ–º —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã
keys_with_multiple_temps = data_temp['key'].value_counts()[data_temp['key'].value_counts() > 1].index
data_temp = data_temp[data_temp['key'].isin(keys_with_multiple_temps)]

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è
plot_numerical_distribution(data_temp, ['temp'])
display(data_temp.describe())
data_temp.info()
````

````python
# –ê–≥—Ä–µ–≥–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö
temp_summary = data_temp.groupby('key').agg(
    temp_start=('temp', 'first'),  # –ü–µ—Ä–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã
    temp_end=('temp', 'last'),     # –ü–æ—Å–ª–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã
    first_temp_measurement=('measurement_time', 'min'),  # –ü–µ—Ä–≤–æ–µ –≤—Ä–µ–º—è –∑–∞–º–µ—Ä–∞
    last_temp_measurement=('measurement_time', 'max')    # –ü–æ—Å–ª–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –∑–∞–º–µ—Ä–∞
).reset_index()

# –î–æ–±–∞–≤–ª—è–µ–º –≤—Ä–µ–º—è –º–µ–∂–¥—É –∑–∞–º–µ—Ä–∞–º–∏ —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã (–≤ –º–∏–Ω—É—Ç–∞—Ö)
temp_summary['temp_measurement_duration'] = (
    temp_summary['last_temp_measurement'] - temp_summary['first_temp_measurement']
).dt.total_seconds() / 60  # –í –º–∏–Ω—É—Ç–∞—Ö

# –ê–Ω–∞–ª–∏–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤ –Ω–æ–≤–æ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ
plot_numerical_distribution(temp_summary, ['temp_start', 'temp_end', 'temp_measurement_duration'])
display(temp_summary.head())
display(temp_summary.describe())
temp_summary.info()
````

–û—Ç–ª–∏—á–Ω–æ, —á—Ç–æ –æ–±—Ä–∞–±–æ—Ç–∞–Ω–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–µ —Å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π —Ä–µ–∞–∫—Ç–∏–≤–Ω–æ–π –º–æ—â–Ω–æ—Å—Ç—å—é

````python
# –î–æ–±–∞–≤–ª—è–µ–º —Å—Ç–æ–ª–±–µ—Ü —Å –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é –Ω–∞–≥—Ä–µ–≤–∞
data_arc['heating_duration'] = (data_arc['end_heating_arc'] - data_arc['start_heating_arc']).dt.total_seconds() / 60 # –í –º–∏–Ω—É—Ç–∞—Ö

# –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
negative_duration_rows = data_arc[data_arc['heating_duration'] < 0]
negative_duration_keys = negative_duration_rows['key'].unique()

# –í—ã–≤–æ–¥–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ –∫–ª—é—á–µ–π —Å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ —Å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é –Ω–∞–≥—Ä–µ–≤–∞: {len(negative_duration_rows)}")
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª—é—á–µ–π —Å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é –Ω–∞–≥—Ä–µ–≤–∞: {len(negative_duration_keys)}")
````

–ê–∫—Ç–∏–≤–Ω–∞—è –º–æ—â–Ω–æ—Å—Ç—å –∏ —Ä–µ–∞–∫—Ç–∏–≤–Ω–∞—è –º–æ—â–Ω–æ—Å—Ç—å –∏–º–µ—é—Ç —Å—Ö–æ–∂–µ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –Ω–∏–∑–∫–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –∏ —É–º–µ—Ä–µ–Ω–Ω–æ–π –≤–∞—Ä–∏–∞—Ç–∏–≤–Ω–æ—Å—Ç—å—é.
–ë–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –∑–Ω–∞—á–µ–Ω–∏–π –º–æ—â–Ω–æ—Å—Ç–∏ —Å–æ—Å—Ä–µ–¥–æ—Ç–æ—á–µ–Ω—ã –≤ —É–∑–∫–æ–º –¥–∏–∞–ø–∞–∑–æ–Ω–µ, —á—Ç–æ –º–æ–∂–µ—Ç –≥–æ–≤–æ—Ä–∏—Ç—å –æ –∫–æ–Ω—Ç—Ä–æ–ª–∏—Ä—É–µ–º–æ–º —Ç–µ—Ö–Ω–æ–ª–æ–≥–∏—á–µ—Å–∫–æ–º –ø—Ä–æ—Ü–µ—Å—Å–µ.
–¢–∞–∫–∂–µ –ø—Ä–æ–≤–µ—Ä–∏–ª–∏ –ª–æ–≥–∏—á–Ω–æ—Å—Ç—å –¥–∞–Ω–Ω—ã—Ö, —á—Ç–æ–± –Ω–∞—á–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ –Ω–µ –ø—Ä–µ–≤—ã—à–∞–ª–∞ –∫–æ–Ω–µ—á–Ω—É—é.

````python
# –ê–≥—Ä–µ–≥–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö –ø–æ –∫–ª—é—á—É (key)
aggregated_data_arc = data_arc.groupby('key').agg({
    'start_heating_arc': ['min'],
    'end_heating_arc': ['max'],
    'active_power': ['count', 'mean', 'sum'],
    'reactive_power': ['mean', 'sum']
}).reset_index()

# –£–ø—Ä–æ—â–µ–Ω–∏–µ –º—É–ª—å—Ç–∏–∏–Ω–¥–µ–∫—Å–∞ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö
aggregated_data_arc.columns = [
    'key', 
    'start_heating_arc', 'end_heating_arc',
    'iteration_count', 
    'active_power_mean', 'active_power_sum',
    'reactive_power_mean', 'reactive_power_sum'
]

# –î–æ–±–∞–≤–ª—è–µ–º –ø–æ–ª–Ω—É—é –º–æ—â–Ω–æ—Å—Ç—å –≤ –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
aggregated_data_arc['total_power_mean'] = (
    aggregated_data_arc['active_power_mean']**2 + aggregated_data_arc['reactive_power_mean']**2
)**0.5

aggregated_data_arc['total_power_sum'] = (
    aggregated_data_arc['active_power_sum']**2 + aggregated_data_arc['reactive_power_sum']**2
)**0.5

# –î–æ–±–∞–≤–ª—è–µ–º –≤—Ä–µ–º—è –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π (—Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –ø–æ—Å–ª–µ–¥–Ω–∏–º –æ–∫–æ–Ω—á–∞–Ω–∏–µ–º –∏ –ø–µ—Ä–≤—ã–º –Ω–∞—á–∞–ª–æ–º)
aggregated_data_arc['heating_duration'] = (
    aggregated_data_arc['end_heating_arc'] - aggregated_data_arc['start_heating_arc']
).dt.total_seconds() / 60  # –í –º–∏–Ω—É—Ç–∞—Ö
````

````python
from matplotlib.gridspec import GridSpec

# –°–ø–∏—Å–æ–∫ —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ (–∏—Å–∫–ª—é—á–∞–µ–º datetime –∏ key)
numerical_columns = [col for col in merged_data.columns if merged_data[col].dtype in ['float64', 'int64']]

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –¥–ª—è —Å—É–±–ø–ª–æ—Ç–æ–≤
num_plots = len(numerical_columns)
num_rows = (num_plots + 1) // 2  # –û–∫—Ä—É–≥–ª—è–µ–º –≤–≤–µ—Ä—Ö

# –°–æ–∑–¥–∞–µ–º —Ñ–∏–≥—É—Ä—É —Å —Å—É–±–ø–ª–æ—Ç–∞–º–∏
fig = plt.figure(figsize=(16, num_rows * 6))
gs = GridSpec(num_rows * 2, 2, height_ratios=[3, 1] * num_rows)  # 3:1 –¥–ª—è –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –∏ –±–æ–∫—Å–ø–ª–æ—Ç–∞

# –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –∫–∞–∂–¥–æ–º—É —Å—Ç–æ–ª–±—Ü—É –∏ —Å—Ç—Ä–æ–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
for i, col in enumerate(numerical_columns):
    row = (i // 2) * 2  # –£–º–Ω–æ–∂–∞–µ–º –Ω–∞ 2, —Ç–∞–∫ –∫–∞–∫ –∫–∞–∂–¥–∞—è –ø–∞—Ä–∞ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞+–±–æ–∫—Å–ø–ª–æ—Ç –∑–∞–Ω–∏–º–∞–µ—Ç 2 —Å—Ç—Ä–æ–∫–∏
    col_idx = i % 2
    
    # –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞
    ax_hist = fig.add_subplot(gs[row, col_idx])
    sns.histplot(merged_data[col].dropna(), bins=30, kde=False, ax=ax_hist, color='blue', edgecolor='black')
    ax_hist.set_title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ {col}', fontsize=12, fontweight='bold')
    ax_hist.set_xlabel(None)
    ax_hist.set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=10)
    ax_hist.grid(axis='y', linestyle='--', alpha=0.5)
    
    # –ë–æ–∫—Å–ø–ª–æ—Ç
    ax_box = fig.add_subplot(gs[row + 1, col_idx])
    sns.boxplot(x=merged_data[col].dropna(), ax=ax_box, color='blue')
    ax_box.set_xlabel(col, fontsize=10)
    ax_box.set_ylabel(None)

# –£–±–∏—Ä–∞–µ–º –ø—É—Å—Ç—ã–µ —Å—É–±–ø–ª–æ—Ç—ã, –µ—Å–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –Ω–µ—á–µ—Ç–Ω–æ–µ
if num_plots % 2 != 0:
    fig.delaxes(fig.add_subplot(gs[-1, -1]))

plt.tight_layout()
plt.show()
````

````python
# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã
correlation_matrix = merged_data.corr()
````

````python
# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ —Ä–∞–∑–º–µ—Ä–∞ –≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(16, 20))

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
sns.heatmap(
    correlation_matrix, 
    annot=True, 
    fmt=".2f", 
    cmap='coolwarm', 
    cbar_kws={"orientation": "horizontal", "shrink": 0.8, "aspect": 40},
    annot_kws={"size": 8}  # –£–º–µ–Ω—å—à–µ–Ω–∏–µ —Ä–∞–∑–º–µ—Ä–∞ —à—Ä–∏—Ñ—Ç–∞ –∞–Ω–Ω–æ—Ç–∞—Ü–∏–π
)

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞
plt.title("Correlation Matrix", fontsize=16)

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ —Ä–∞–∑–º–µ—Ä–∞ —à—Ä–∏—Ñ—Ç–∞ –º–µ—Ç–æ–∫ –æ—Å–µ–π
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plt.show()
````

````python
# –†–∞–∑–¥–µ–ª—è–µ–º —Å —É—á–µ—Ç–æ–º —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π temp_end
train_data, test_data = train_test_split(data, test_size=0.25, random_state=RANDOM_STATE)
````

````python
# –£–¥–∞–ª—è–µ–º –∞–Ω–æ–º–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ —Å –ø–æ–º–æ—à—å—é –ø–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞ –∫–ª—é—á–µ–π
train_data = train_data[~train_data.index.isin(keys_to_remove)]

# –í—ã–≤–æ–¥–∏–º —Ä–∞–∑–º–µ—Ä—ã –∏—Ç–æ–≥–æ–≤—ã—Ö –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–æ–≤
print(f"–†–∞–∑–º–µ—Ä —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏: {train_data.shape}")
print(f"–†–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏: {test_data.shape}")
````

````python
X_train = train_data.drop('temp_end', axis=True)
y_train = train_data['temp_end']
X_test = test_data.drop('temp_end', axis=True)
y_test = test_data['temp_end']
````

### –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

–î–ª—è –∑–∞–¥–∞—á–∏ –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã (`temp_end`) –Ω–∞ –æ—Å–Ω–æ–≤–µ –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –º–æ–∂–Ω–æ —Ä–∞—Å—Å–º–æ—Ç—Ä–µ—Ç—å –¥–≤–µ –º–æ–¥–µ–ª–∏: –æ–¥–Ω—É –∫–ª–∞—Å—Å–∏—á–µ—Å–∫—É—é –∏ –æ–¥–Ω—É —Å–æ–≤—Ä–µ–º–µ–Ω–Ω—É—é. –í –∫–∞—á–µ—Å—Ç–≤–µ –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–æ–π –º–æ–¥–µ–ª–∏ –∏—Å–ø–æ–ª—å–∑—É–µ–º –ª–∏–Ω–µ–π–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é —Å —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–µ–π - `Ridge`, –∞ –≤ –∫–∞—á–µ—Å—Ç–≤–µ —Å–ª–æ–∂–Ω–æ–π –º–æ–¥–µ–ª–∏ –º–æ–¥–µ–ª—å `LightGBM`.

````python
robust_scaler = RobustScaler()
X_train_robust = robust_scaler.fit_transform(X_train)
X_test_robust = robust_scaler.transform(X_test)

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –º–∞—Å—Å–∏–≤ –æ–±—Ä–∞—Ç–Ω–æ –≤ DataFrame
X_train_robust = pd.DataFrame(X_train_robust, columns=X_train.columns)
X_test_robust = pd.DataFrame(X_test_robust, columns=X_test.columns)
````

#### Ridge

````python
ridge_model = Ridge(random_state=RANDOM_STATE)
ridge_params = {
    'alpha': [0.01, 0.1, 1, 10, 100],  # –°–∏–ª–∞ —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏
    'fit_intercept': [True, False]
}


ridge_search = RandomizedSearchCV(
    ridge_model, ridge_params, n_iter=10, cv=5, scoring='neg_mean_absolute_error', random_state=RANDOM_STATE
)
````
